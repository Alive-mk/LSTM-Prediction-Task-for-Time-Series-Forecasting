# 基于LSTM的FRED时序数据集预测任务

实验内容：根据所给的美国公用事业电力和天然气工业生产的产量数据，对2022年9月至2023年8月的产量进行预测。

数据集内容：1996年8月至2023年8月时间段美国公用事业部门电力和天然气工业生产的产量数据，共325条。

数据集来源：Federal Reserve Economic Database (FRED)，网址https://fred.stlouisfed.org/series/IPG2211A2N

本实验的主要步骤有：
* 导入相关库和数据集
* 数据预处理
* 模型建立
* 模型训练
* 模型评估


## 一.导入相关库、数据集
导入相关库，导入数据集并打印查看数据。

```python
import torch
import torch.nn as nn
from sklearn.preprocessing import MinMaxScaler

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

df = pd.read_csv('Dataset_EnergyProduction.csv',index_col=0,parse_dates=True)
df.dropna(inplace=True)#删除DataFrame中包含缺失值NaN的行，inplace参数表示在原始DataFrame上进行操作，而不是创建一个新DF
df=df[691:]#选取最新的325个数据
print(len(df))#打印数据集的样本数量
df.head()#显示DataFrame的前几行数据，默认显示前5行，方便查看数据的样式和内容

df = pd.read_csv('Dataset_EnergyProduction.csv',index_col=0,parse_dates=True)
df.dropna(inplace=True)
df=df[691:]
print(len(df))
df.head()
```

```python
# 查看数据集
plt.figure(figsize=(12,4))
plt.title('Industrial Production Index for Electricity and Gas Utilities')
plt.ylabel('Index 2017=100, Not Seasonally Adjusted')
plt.grid(True) #网格线
plt.autoscale(axis='x',tight=True)
plt.plot(df['IPG2211A2N'])#列名
plt.show()
```

## 二. 数据预处理
这一步骤主要包括：
* 划分训练集和测试集
* 数据规范化
* 加工数据格式为序列

### 1. 划分训练集和测试集
已知所使用的数据集数量为325，按照313：12划分为训练集和测试集。
设置滑动窗口长度为12。


```python
y = df['IPG2211A2N'].values.astype(float)

test_size = 12
window_size = 12

train_set = y[:-test_size]
test_set = y[-test_size:]

print(f'Train: {len(train_set)}')
print(f'Test:  {len(test_set)}')
```

```python
train_set.reshape(-1,1)
```

### 2. 数据规范化处理
将数据取值范围映射在[-1,1]之间。

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
#fit_transform传入scaler的要是一列数据，而不是一行
train_norm = scaler.fit_transform(train_set.reshape(-1, 1))

print(f'First item, original: {train_set[0]}')
print(f'First item, scaled: {train_norm[0]}')

#查看规范化后的训练集
train_norm
```

### 3. 处理输入模型的数据对象
按照设定的滑动窗口大小window_size，将有时序的数据组合成序列，给LSTM模型读入。

```python
train_norm = torch.FloatTensor(train_norm).view(-1)

def input_data(seq,ws):
    out = []
    L = len(seq)
    for i in range(L-ws):
        window = seq[i:i+ws]
        label = seq[i+ws:i+ws+1]
        out.append((window,label))
    return out

#train_data内包含了若干个tuple的list。tuple元素也为list
train_data = input_data(train_norm,window_size)

#最终输入模型的训练数据长度为313-12=301
print(f'Train_data: {len(train_data)}')
```

## 三. 模型建立
### 1. 定义LSTM网络结构
添加一个 (1,64) LSTM layer ，添加一个(64,1) 全连接层(线性Linear)。

初始化参数$h_0$ and $c_0$。

前向传播函数只需要返回最后一个预测值。

```python
class LSTMnetwork(nn.Module):
    def __init__(self,input_size=1,hidden_size=64,output_size=1):
        super().__init__()
        self.hidden_size = hidden_size
        
        # 添加一个 LSTM 层:
        self.lstm = nn.LSTM(input_size,hidden_size)
        
        # 添加一个全连接层 fully-connected layer:
        self.linear = nn.Linear(hidden_size,output_size)
        
        # 初始化参数 h0 and c0:
        self.hidden = (torch.zeros(1,1,self.hidden_size),
                       torch.zeros(1,1,self.hidden_size))

    def forward(self,seq):
        lstm_out, self.hidden = self.lstm(
            seq.view(len(seq),1,-1), self.hidden)
        pred = self.linear(lstm_out.view(len(seq),-1))
        return pred[-1]
```

```python
#查看参数的维度、形状，理解各层的输入
torch.tensor(range(12)).view(12,1,-1)
```

### 2. 初始化模型
该步骤包括网络结构实体化、定义损失函数criterion、定义优化器optimizer。

这里选用的损失函数为MSELoss，选用的优化器是学习率为0.001的Adam。

```python
torch.manual_seed(101)

#网络结构实体化
model = LSTMnetwork()
#损失函数
criterion = nn.MSELoss()
#优化器 torch.optim模块
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

model
```

## 四. 模型训练
在模型训练过程中，主要步骤有：前向传播、损失函数计算、反向传播、优化器更新模型权重。

本实验模型设置的训练迭代次数为50。

每次迭代前重置隐藏状态：
* 防止信息泄露。
在每个训练样本之间，重新初始化隐藏状态可以防止来自不同样本的信息相互干扰。如果在不同样本之间保留旧的隐藏状态，那么可能会导致模型在处理不同样本时混淆信息，这会影响训练的稳定性和准确性。
* 防止梯度消失和梯度爆炸。
梯度通常通过时间反向传播（Backpropagation Through Time, BPTT）进行传递。如果在训练中不定期重置隐藏状态，梯度可能会在时间上传播得非常远，导致梯度消失或梯度爆炸问题，使训练不稳定，模型收敛困难。


```python
epochs = 50

for i in range(epochs):
    for seq, y_train in train_data:
        
        # reset parameters手动梯度清零
        optimizer.zero_grad()
        #重置隐藏状态reset hidden state-----时序预测实验易错步骤
        model.hidden = (torch.zeros(1,1,model.hidden_size),
                        torch.zeros(1,1,model.hidden_size))
        
        # forward
        y_pred = model(seq)
        # loss function
        loss = criterion(y_pred, y_train)
        #backward反向计算出各参数的梯度
        loss.backward()
        #更新网络中的参数
        optimizer.step()

    # print statement
    print(f'{i+1} of {epochs} epochs completed')
```

## 五. 模型评估和保存
### 1.模型评估
在使用模型进行预测之前，对参数$h_0$ 和 $c_0$ 重新初始化。

接着，使用inverse_transform()函数将模型预测的值映射为原数据集的取值范围。

```python
#从nparray转化为list
preds = train_norm[-window_size:].tolist()

#切换为评估模式：测试时停用 Batch Normalization 和 Dropout，使用训练好的值
model.eval()

future = 12
for i in range(future):
    seq = torch.FloatTensor(preds[-window_size:])
    with torch.no_grad():#预测时不用自动求导
        model.hidden = (torch.zeros(1,1,model.hidden_size),
                        torch.zeros(1,1,model.hidden_size))
        preds.append(model(seq).item())
        #<tensor>.item()获取tensor内的元素（非tensor类型）
        
preds[window_size:]
```

```python
#invere_transform的输入数据也是一'列'
true_predictions = scaler.inverse_transform(np.array(preds[window_size:]).reshape(-1, 1))
true_predictions
```

将模型预测的结果与初始数据对比进行可视化，用折线图表示。

```python
x = np.arange('2022-09-01', '2023-09-01', dtype='datetime64[M]').astype('datetime64[D]')

plt.figure(figsize=(12,4))
plt.title('Industrial Production Index for Electricity and Gas Utilities')
plt.ylabel('Index 2017=100, Not Seasonally Adjusted')
plt.grid(True)
plt.autoscale(axis='x',tight=True)
plt.plot(df['IPG2211A2N'])
plt.plot(x,true_predictions)
plt.show()
```
